### 2 - Realizar un **script básico** que calcule las ventas mensuales utilizando variables y operadores.

```python
# Modifico precio y fecha que estan como objeto a numero y fecha respectivamente, previo a eliminar el simbolo $ en precio
df_ven['precio'] = df_ven['precio'].astype(str).str.replace('$', '', regex= False)
df_ven['precio'] = pd.to_numeric(df_ven['precio'], errors='coerce')
df_ven['fecha_venta'] = pd.to_datetime(df_ven['fecha_venta'], format='%d/%m/%Y')
```

```python
# Calculo la venta total por cada transacción (precio * cantidad) usando operadores.
df_ven['venta_total_transaccion'] = df_ven['precio'] * df_ven['cantidad']

# Extraigo el mes de la fecha de venta.
df_ven['mes'] = df_ven['fecha_venta'].dt.month

# Agrupo por mes y sumo las ventas totales para obtener las ventas mensuales.
ventas_mensuales = df_ven.groupby('mes')['venta_total_transaccion'].sum()

# Mapeo de números de mes a nombres de mes para una mejor lectura
meses_nombres = {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio',
    7: 'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
}

print("Ventas mensuales:")
for mes_num, total_venta in ventas_mensuales.items():
    print(f"  {meses_nombres[mes_num]}: ${total_venta:,.2f}")
```

### 2 - **Transformación de Datos:** Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.

```python
print("Creando la tabla de productos de alto rendimiento\n")

# 1. Calculo las ventas totales por producto

ventas_por_producto = df_ven.groupby('producto')['venta_total_transaccion'].sum().reset_index()
ventas_por_producto = ventas_por_producto.rename(columns={'venta_total_transaccion': 'total_ventas'})

# 2. Defino el umbral de alto rendimiento

umbral_alto_rendimiento = ventas_por_producto['total_ventas'].quantile(0.75)

print(f"El umbral para considerar un producto de 'alto rendimiento' (percentil 75) es: ${umbral_alto_rendimiento:,.2f}\n")

# 3. Filtro los productos que superan el umbral

productos_alto_rendimiento = ventas_por_producto[ventas_por_producto['total_ventas'] >= umbral_alto_rendimiento]

# 4. Ordeno los productos por ventas totales de mayor a menor

productos_alto_rendimiento = productos_alto_rendimiento.sort_values(by='total_ventas', ascending=False)

# 5. Muestro la tabla resultante

print("Tabla de productos con Alto Rendimiento:\n")
display(productos_alto_rendimiento)

print(f"\nSe han identificado {len(productos_alto_rendimiento)} productos de alto rendimiento.")
```

## Etapa 3: Análisis de Datos


### 1 - Estadística Descriptiva

Se calculan medidas de tendencia central (media, mediana, moda) y de dispersión (desviación estándar, rango intercuartílico) sobre las variables numéricas de ventas: `precio`, `cantidad` y `venta_total_transaccion`.

```python
# Calculamos las medidas de tendencia central y dispersión para las variables numéricas de ventas

columnas_interes = ['precio', 'cantidad', 'venta_total_transaccion']
resumen_estadistico = {}

for col in columnas_interes:
    media = df_ven[col].mean()
    mediana = df_ven[col].median()
    moda = df_ven[col].mode()[0]
    desv_std = df_ven[col].std()
    q1 = df_ven[col].quantile(0.25)
    q3 = df_ven[col].quantile(0.75)
    iqr = q3 - q1

    resumen_estadistico[col] = {
        'Media': media,
        'Mediana': mediana,
        'Moda': moda,
        'Desv. Estándar': desv_std,
        'Q1': q1,
        'Q3': q3,
        'Rango Intercuartílico (IQR)': iqr
    }

df_resumen_estadistico = pd.DataFrame(resumen_estadistico).T
print("Resumen de Estadística Descriptiva - Ventas\n")
display(df_resumen_estadistico.round(2))
```

**Lectura de los resultados:**
- El `precio` promedio de los productos es de aproximadamente $75, con una dispersión moderada (desvío ~$29).
- La `cantidad` vendida por transacción tiene una media cercana a 6-7 unidades.
- La `venta_total_transaccion` muestra mayor dispersión relativa (desvío ~$334), lo cual es esperable al ser el producto de dos variables (precio x cantidad). La diferencia entre media y mediana sugiere una distribución con cierta asimetría hacia la derecha (algunas ventas puntuales muy altas).

### 2 - EDA (Análisis Exploratorio de Datos)

Se analiza la distribución de las ventas, se identifican posibles outliers mediante el criterio de rango intercuartílico (IQR) y se observa la tendencia mensual y por categoría de las ventas.

```python
# Extraemos el mes para poder analizar la tendencia temporal de las ventas
df_ven['mes'] = df_ven['fecha_venta'].dt.month

# Detección de outliers en venta_total_transaccion usando el criterio de IQR
q1 = df_ven['venta_total_transaccion'].quantile(0.25)
q3 = df_ven['venta_total_transaccion'].quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

outliers = df_ven[(df_ven['venta_total_transaccion'] < limite_inferior) | (df_ven['venta_total_transaccion'] > limite_superior)]

print(f"Límites para considerar un valor atípico (outlier): [{limite_inferior:.2f}, {limite_superior:.2f}]")
print(f"Cantidad de transacciones consideradas outliers: {len(outliers)} de {len(df_ven)} ({len(outliers)/len(df_ven)*100:.2f}%)\n")
display(outliers[['producto', 'precio', 'cantidad', 'venta_total_transaccion', 'categoria']].sort_values(by='venta_total_transaccion', ascending=False))
```

```python
# Tendencia mensual de ventas: suma, promedio y cantidad de transacciones por mes
tendencia_mensual = df_ven.groupby('mes')['venta_total_transaccion'].agg(
    ventas_totales='sum',
    ticket_promedio='mean',
    cantidad_transacciones='count'
).reset_index()

meses_nombres = {1:'Enero',2:'Febrero',3:'Marzo',4:'Abril',5:'Mayo',6:'Junio',
                 7:'Julio',8:'Agosto',9:'Septiembre',10:'Octubre',11:'Noviembre',12:'Diciembre'}
tendencia_mensual['mes_nombre'] = tendencia_mensual['mes'].map(meses_nombres)

print("Tendencia mensual de ventas:\n")
display(tendencia_mensual[['mes_nombre','ventas_totales','ticket_promedio','cantidad_transacciones']].round(2))

plt.figure(figsize=(10,4))
sns.lineplot(data=tendencia_mensual, x='mes', y='ventas_totales', marker='o', color='#3b6e8f')
plt.xticks(tendencia_mensual['mes'], [meses_nombres[m][:3] for m in tendencia_mensual['mes']])
plt.title('Tendencia Mensual de Ventas Totales')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales ($)')
plt.tight_layout()
plt.show()
```

**Hallazgos del EDA:**
- Se detectó un porcentaje bajo de outliers en `venta_total_transaccion`, asociados a transacciones con cantidades y/o precios altos en simultáneo.
- Las ventas totales se mantienen relativamente estables a lo largo del año, sin una estacionalidad muy marcada, aunque se observan picos en marzo y mayo, y una leve baja hacia mitad de año.

### 3 - Correlación entre Variables

Se evalúa la relación entre `precio`, `cantidad` y `venta_total_transaccion` mediante una matriz de correlación, visualizada como mapa de calor.

```python
# Calculamos la matriz de correlación entre las variables numéricas de interés
matriz_correlacion = df_ven[['precio', 'cantidad', 'venta_total_transaccion']].corr()

print("Matriz de correlación:\n")
display(matriz_correlacion.round(3))

plt.figure(figsize=(6,5))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Mapa de Calor - Correlación entre Variables de Venta')
plt.tight_layout()
plt.show()
```

**Lectura de la correlación:**
- `precio` y `cantidad` prácticamente no están correlacionados (~0), lo que indica que el precio de un producto no influye en la cantidad que se compra por transacción (no hay un efecto de "a mayor precio, menor cantidad" en este dataset).
- `cantidad` tiene una correlación fuerte y positiva con `venta_total_transaccion` (~0.77), siendo el principal motor del monto vendido.
- `precio` también correlaciona positivamente con `venta_total_transaccion` (~0.56), aunque en menor medida que la cantidad.

### 4 - Consolidación de Datos: Ventas dentro vs. fuera de Campaña

Se integra el dataset de `ventas` con el de `marketing` para evaluar si las campañas de marketing tuvieron un impacto observable en las ventas. El enfoque elegido es: **para cada venta, determinar si su fecha cayó dentro del rango de alguna campaña activa para ese mismo producto**, y comparar las ventas dentro vs. fuera de campaña.

```python
# Unimos ventas y marketing por producto (cada venta se cruza con todas las campañas de ese producto)
df_ventas_marketing = pd.merge(df_ven, df_mar, on='producto', how='inner', suffixes=('_venta', '_marketing'))

# Marcamos si la fecha de la venta cae dentro del rango de esa campaña específica
df_ventas_marketing['en_campana'] = (
    (df_ventas_marketing['fecha_venta'] >= df_ventas_marketing['fecha_inicio']) &
    (df_ventas_marketing['fecha_venta'] <= df_ventas_marketing['fecha_fin'])
)

print("DataFrame consolidado de Ventas y Marketing:")
display(df_ventas_marketing.head())
```

```python
# Como cada venta puede cruzarse con varias campañas del mismo producto, consolidamos a nivel de venta:
# una venta queda marcada 'en campaña' si coincidió con AL MENOS UNA campaña activa
flag_campana_por_venta = df_ventas_marketing.groupby('id_venta')['en_campana'].any().reset_index()

df_consolidado = df_ven.merge(flag_campana_por_venta, on='id_venta', how='left')
df_consolidado['en_campana'] = df_consolidado['en_campana'].fillna(False)
df_consolidado['estado_campana'] = df_consolidado['en_campana'].map({True: 'Dentro de campaña', False: 'Fuera de campaña'})

print("Distribución de transacciones dentro vs fuera de campaña:")
display(df_consolidado['estado_campana'].value_counts().to_frame(name='Cantidad de transacciones'))
```

```python
# Comparamos métricas clave de ventas dentro vs fuera de campaña
comparacion_campana = df_consolidado.groupby('estado_campana').agg(
    ventas_totales=('venta_total_transaccion', 'sum'),
    ticket_promedio=('venta_total_transaccion', 'mean'),
    cantidad_promedio=('cantidad', 'mean'),
    n_transacciones=('id_venta', 'count')
).reset_index()

print("Comparación de ventas: Dentro vs Fuera de Campaña\n")
display(comparacion_campana.round(2))
```

```python
# Ventas por categoría, separando dentro y fuera de campaña
ventas_categoria_campana = df_consolidado.groupby(['categoria', 'estado_campana'])['venta_total_transaccion'].sum().reset_index()

print("Ventas por Categoría según estado de campaña:\n")
display(ventas_categoria_campana.round(2))
```

```python
# Análisis adicional: efectividad por canal de marketing (ventas generadas en campaña vs costo de la campaña)
ventas_en_campana = df_ventas_marketing[df_ventas_marketing['en_campana']]

efectividad_canal = ventas_en_campana.groupby('canal').agg(
    ventas_en_campana=('venta_total_transaccion', 'sum'),
    costo_total=('costo', 'sum')
).reset_index()
efectividad_canal['retorno_aprox'] = (efectividad_canal['ventas_en_campana'] / efectividad_canal['costo_total']).round(2)

print("Efectividad aproximada por canal de marketing (ventas en campaña / costo de campaña):\n")
display(efectividad_canal.sort_values(by='retorno_aprox', ascending=False))
```

**Insights de la consolidación:**
- El volumen de ventas "dentro de campaña" es bastante menor en cantidad de transacciones (las campañas cubren ventanas de tiempo acotadas), pero el **ticket promedio y la cantidad promedio por transacción son prácticamente iguales** dentro y fuera de campaña. Esto sugiere que, en este dataset, las campañas no estarían generando un incremento en el monto promedio de compra, sino que su efecto (si existe) habría que buscarlo en otra dimensión, como el volumen total de transacciones.
- Los tres canales de marketing (Email, TV, RRSS) muestran una efectividad aproximada (ventas generadas sobre costo invertido) relativamente pareja, con Email y TV levemente por encima de RRSS.

- Los histogramas de `precio` y `cantidad` muestran distribuciones bastante uniformes, sin una concentración marcada en ningún valor particular.
- El histograma de `venta_total_transaccion` muestra una cola hacia la derecha (asimetría positiva), consistente con la diferencia entre media y mediana observada en la estadística descriptiva.
- Los boxplots confirman la presencia de algunos outliers en `venta_total_transaccion` (puntos por fuera de los bigotes), mientras que `precio` y `cantidad` no presentan outliers significativos.
